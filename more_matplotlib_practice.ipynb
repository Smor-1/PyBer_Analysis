```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Resources/PyBer_ride_data.csv")
```

```python
df.head(2)
```

```python
# setting x axis and tick locations
x_axis = np.arange(len(df))
tick_locations = [value for value in x_axis]

df.plot(x = "Month" , y = "Avg. Fare ($USD)")
plt.xticks(tick_locations, df["Month"])
plt.show()
```

```python
df.plot.bar(x = "Month" , y = "Avg. Fare ($USD)")
plt.show()
```

```python
#can also do this

df.plot(x = "Month" , y = "Avg. Fare ($USD)" , kind = "bar")
plt.show()
```
