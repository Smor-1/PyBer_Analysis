```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

#the % is sometimes called the magic command - running a cell with it at the start the "magic" happens behind the scenes
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis , y_axis)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis , y_axis)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
#width, height
plt.figure(figsize = (7, 5))
plt.plot(x_axis , y_axis , label = "Boston" , marker = "*" , color = "orange" , linewidth = 0.5)
#min, max, can use string or index position!! 
plt.xlim("Jan",12)
plt.ylim(9.5,45)
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
#way to add grid lines below
#plt.grid()
plt.savefig("FIGURE NAME")
plt.show()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.bar(x_axis , y_axis , label = "Boston" , color = "red")
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.legend()
plt.show()
```

```python
#using barh for horizontal
plt.barh(x_axis , y_axis , label = "Boston" , color = "magenta")

#inverting to get Jan at the top, use gca (get current axis) and invert_yaxis()
plt.gca().invert_yaxis()

plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.legend()

plt.show()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.scatter(y_axis, x_axis , color = "red" , label = "Chicago")
#OR can use: plt.plot(x_axis, y_axis, 'o')

plt.gca().invert_yaxis()

plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.legend()

plt.show()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#because the small markers look super small so lets * everything by 3 to make it bigger
y_axis_times_3 = []
for data in y_axis:
    y_axis_times_3.append(data*3)

#s for size of the value in the y axis to make the markers bigger for larger values! 
plt.scatter(x_axis, y_axis, s=y_axis_times_3)

plt.show()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]


#list comprehension technique below to get the same as above
plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis])

plt.show()
```

```python
#length of tuple equal to the number of wedges in the pie chart (12)
explode_values = (0,0.2,0.2,0,0.5,0,0,0,0.5,0,0,0)

#12 colours, 1 for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.figure(figsize = (8,8))

plt.pie(y_axis , colors = colors , labels = x_axis , explode = explode_values , autopct='%.1f%%')

plt.show()
```
