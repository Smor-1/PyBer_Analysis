```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
#can also say %matplotlib notebook to make it interactive!!!! 
%matplotlib inline
import numpy as np
```

```python
x_axis = np.arange(0,8,0.1)
```

```python
exponent = [np.exp(x) for x in x_axis]
```

```python
#exponent
```

```python
plt.plot(x_axis , exponent)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

```python
sin = np.sin(x_axis)
cos = np.cos(x_axis)
plt.plot(x_axis, sin , color = 'red')
plt.plot(x_axis, cos , color = "green")
plt.show()
```
